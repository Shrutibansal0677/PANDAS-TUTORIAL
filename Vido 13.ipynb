# CROSSTAB

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('C:/Users/HP/Downloads/survey.xls')
df
```

## crosstab

```python
pd.crosstab(df.Sex, df.Handedness)    #frequency distribution of male female being left handed or right handed
```

```python
pd.crosstab(df.Sex, df.Nationality)     #frequency distribution of male female belonging to which nationality
```

## margins

```python
pd.crosstab(df.Sex, df.Nationality, margins=True)  
```

## multiple rows and index 

```python
pd.crosstab([df.Sex, df.Handedness], df.Nationality, margins=True)  
```

```python
pd.crosstab(df.Sex, [df.Handedness, df.Nationality], margins=True)  
```

## normalize

```python
pd.crosstab(df.Sex, [df.Handedness, df.Nationality],  normalize='index')  
```

```python
pd.crosstab(df.Sex, [df.Handedness, df.Nationality],  margins= True, normalize='index')  
```

```python
pd.crosstab([df.Sex, df.Handedness], df.Nationality, margins=True, normalize='index')  
```

```python
pd.crosstab([df.Sex, df.Handedness], df.Nationality, margins=True, normalize='columns')  
```

## aggfunc and values

```python
pd.crosstab(df.Nationality, df.Sex, values=df.Age, aggfunc=np.average)
```
