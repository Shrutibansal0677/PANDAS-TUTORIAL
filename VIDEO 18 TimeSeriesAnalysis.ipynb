```python
# creating own calendar

from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday

class mycal(AbstractHolidayCalendar):
    
    rules=[
        Holiday('My chutti', month=1, day=1)
    ]
```

```python
mycalendar= CustomBusinessDay(calendar= mycal())
```

```python
rng1= pd.date_range(start='12/30/2024', periods= 22, freq=mycalendar)
```

```python
#to set customized freemask or working days

myworking= CustomBusinessDay(weekmask='Sun Mon Tue Wed Thu', holidays=['2025-01-23'])
rng2= pd.date_range(start='12/30/2024', periods= 22, freq=myworking)
rng2
```
