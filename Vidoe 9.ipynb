# MERGE FUNCTION 

```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({
    'city':['ny', 'nd', 'wc', 'wt'],
    'temp':[34,45,43,23]
})
df1
```

```python
df2=pd.DataFrame({
    'city':['ny', 'wt', 'wc', 'nd'],
    'humidity':[12, 21, 11, 10]
})
df2
```

```python
df3=pd.merge(df1, df2, on="city")
df3
```

```python
df3=pd.merge(df1, df2, on="city" , indicator='True')
df3
```

```python
df4=pd.DataFrame({
    'city':['ny', 'wt', 'wc', 'nd'],
    'humidity':[12, 21, 11, 10]
})
df4
```

```python
df5=pd.DataFrame({
    'city':['ny', 'wt', 'wc'],
    'humidity':[12, 21, 11]
})
df5
```

```python
df6=pd.merge(df4, df5, on="city", suffixes=['left', 'right'])
df6
```

```python
df6=pd.merge(df4, df5, on="city", how='outer', suffixes=['left', 'right'])
df6
```

```python
df6=pd.merge(df4, df5, on="city", how='left', suffixes=['left', 'right'])
df6
```

```python
df6=pd.merge(df4, df5, on="city", how='right', suffixes=['left', 'right'])
df6
```
