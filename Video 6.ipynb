# Replace function: handling missing values


```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:/Users/HP/Downloads/weather_data (1).csv")
```

```python
df
```

## 1. replacing single value with NaN

```python
df.replace('-99999', np.NaN)
```

```python
df.replace(-99999, np.NaN)
```

```python
df.replace(-99999, np.NaN, inplace=True)
```

## replacing list of values with NaN

```python

df.replace([-99999, '0'], np.NaN)
```

```python
df
```

## replacing values column wise


```python
df.replace({
    'temperature': -99999 ,
    'windspeed' : [-99999,2], 
    'event' : '0'},
    np.NaN
)
```

## replacing values using dictionary

```python
df.replace({'temperature': -99999 ,
    'windspeed' : [-99999,2], 
    'event' : '0'}, 
    [0,1,2]
    )
```

```python
df.replace({
    -99999 : np.NaN ,
    '0' : 'No event'
})
```

## replacing regex with blank values

```python
df.replace('[A-Za-z]', '1', regex=True)
```

```python
df.replace('[A-Za-z]', '', regex=True)
```

## replacing regex with blank values column wisee

```python
df.replace({
    'event' : '[A-Za-z]'
}, ' ', regex=True)
```

## replacing list of values with list of values

```python
df.replace([-99999, '0'], [np.NaN, 'No event'])
```
